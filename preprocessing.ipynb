```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# ------------------------------------------------------
# ğŸ“‚ 0. ë°ì´í„° ë¡œë“œ & ì´ˆê¸° ì •ë¦¬
# ------------------------------------------------------
data_path = 'data/datasets.csv'
data = pd.read_csv(data_path)
```

```python
# ë¶ˆí•„ìš”í•œ ì»¬ëŸ¼ ì œê±°
drop_cols = ['EmployeeNumber', 'EmployeeCount', 'Over18', 'StandardHours']
data.drop(columns=drop_cols, inplace=True)
```

```python
# íƒ€ê²Ÿê°’ ë³€í™˜:ê° ë²”ì£¼í˜• ë³€ìˆ˜ ìˆ˜ì¹˜ ë§¤í•‘
data['Attrition'] = data['Attrition'].map({
    'Yes': 1, 
    'No': 0
})

data['BUSINESS TRAVEL'] = data['BUSINESS TRAVEL'].map({
    'No Travel':1, 
    'Travel Frequently':2, 
    'Tavel Rarely':3
})

data['DEPARTMENT'] = data['DEPARTMENT'].map({
    'HR':1,
    'R&D':2,
    'Sales':3
})

data['EDUCATION_FIELD'] = data['EDUCATION_FIELD'].map({
    'HR': 1,
    'Life Sciences': 2,
    'Marketing': 3,
    'Medical': 4,
    'Other': 5,
    'Technical Degree': 6
})

data['GENDER'] = data['GENDER'].map({
    'Female': 1,
    'Male': 2
})

data['JOB_ROLE'] = data['JOB_ROLE'].map({
    'Healthcare Representative': 1,
    'Human Resources': 2,
    'Laboratory Technician': 3,
    'Manager': 4,
    'Managing Director': 5,
    'Research Director': 6,
    'Research Scientist': 7,
    'Sales Executive': 8,
    'Sales Representative': 9
})

data['MARITAL_STATUS'] = data['MARITAL_STATUS'].map({
    'Divorced': 1,
    'Married': 2,
    'Single': 3
})
```

```python
# ------------------------------------------------------
# âš ï¸ 1. ì´ìƒê°’ ë° ê²°ì¸¡ ì²˜ë¦¬
# ------------------------------------------------------
# 0ìœ¼ë¡œ ë‚˜ëˆ„ê¸° ë°©ì§€ìš© NaN ì²˜ë¦¬
data['TotalWorkingYears'] = data['TotalWorkingYears'].replace(0, np.nan)
data['YearsAtCompany'] = data['YearsAtCompany'].replace(0, np.nan)
```

```python
# ------------------------------------------------------
# ğŸ¯ 2. ì»¤ë¦¬ì–´ ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜
# ------------------------------------------------------
data['CareerStartAge'] = data['Age'] - data['TotalWorkingYears']                  # ì²« ì»¤ë¦¬ì–´ ì‹œì‘ ë‚˜ì´
data['LoyaltyRatio'] = data['YearsAtCompany'] / data['TotalWorkingYears']        # ì¶©ì„±ë„ ë¹„ìœ¨
data['PromotionDelay'] = data['YearsAtCompany'] - data['YearsSinceLastPromotion']# ìŠ¹ì§„ê¹Œì§€ ì†Œìš” ê¸°ê°„
```

```python
# ------------------------------------------------------
# ğŸ¢ 3. ê·¼ì† ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜
# ------------------------------------------------------
data['CompanyLoyalty'] = data['YearsAtCompany'] / data['TotalWorkingYears']      # íšŒì‚¬ ì „ì²´ ê·¼ì† ë¹„ìœ¨
data['RoleStability'] = data['YearsInCurrentRole'] / data['YearsAtCompany']      # í˜„ì¬ ì§ë¬´ ì§€ì† ë¹„ìœ¨
data['ManagerBond'] = data['YearsWithCurrManager'] / data['YearsAtCompany']      # ë§¤ë‹ˆì € ìœ ëŒ€ ë¹„ìœ¨
```

```python
# ------------------------------------------------------
# ğŸ” 4. ì´ì§/ìŠ¹ì§„ ê²½í–¥
# ------------------------------------------------------
data['JobHopFrequency'] = data['NumCompaniesWorked'] / data['TotalWorkingYears'] # ì—°í‰ê·  ì´ì§ íšŸìˆ˜
data['RecentlyPromoted'] = (data['YearsSinceLastPromotion'] <= 3).astype(int)     # ìµœê·¼ 3ë…„ ë‚´ ìŠ¹ì§„ ì—¬ë¶€
```

```python
# ------------------------------------------------------
# ğŸ’° 5. ê¸‰ì—¬ ë° ì„±ê³¼ ê´€ë ¨ ë³€ìˆ˜
# ------------------------------------------------------
data['IncomePerYear'] = data['MonthlyIncome'] / data['TotalWorkingYears']        # ê²½ë ¥ ëŒ€ë¹„ ì†Œë“
data['HighPerformerQuit'] = ((data['PerformanceRating'] == 4) & 
                             (data['Attrition'] == 1)).astype(int)                # ê³ ì„±ê³¼ ì´íƒˆì ì—¬ë¶€
```

```python
# ------------------------------------------------------
# ğŸ›£ï¸ 6. ìŠ¤íŠ¸ë ˆìŠ¤ ë° ì´ë™ ê´€ë ¨ ë³€ìˆ˜
# ------------------------------------------------------
data['LongCommute'] = (data['DistanceFromHome'] > 20).astype(int)                # ì¥ê±°ë¦¬ ì¶œí‡´ê·¼ ì—¬ë¶€
data['OverworkedPoorBalance'] = ((data['OverTime'] == 'Yes') & 
                                 (data['WorkLifeBalance'] <= 2)).astype(int)      # ì›Œë¼ë°¸ ë¶•ê´´ ì—¬ë¶€
```

```python
# ------------------------------------------------------
# ğŸ˜Š 7. ì¢…í•© ë§Œì¡±ë„
# ------------------------------------------------------
data['OverallSatisfaction'] = (
    data['EnvironmentSatisfaction'] + 
    data['JobSatisfaction'] + 
    data['RelationshipSatisfaction']
) / 3                                                                             # í‰ê·  ë§Œì¡±ë„
```

```python
# ------------------------------------------------------
# ğŸ§‘â€ğŸ’¼ 8. ì§ê¸‰ ê¸°ë°˜ ê³ ìœ„ì§ êµ¬ë¶„
# ------------------------------------------------------
manager_roles = ['Manager', 'Research Director']
data['Is_Manager'] = ((data['JobRole'].isin(manager_roles)) & 
                      (data['JobLevel'] >= 3)).astype(int)                        # ê³ ìœ„ì§ ì—¬ë¶€
```

```python
# ------------------------------------------------------
# ğŸ§¹ 9. ê²°ì¸¡ê°’ ì •ë¦¬
# ------------------------------------------------------
data.fillna(0, inplace=True)
```

```python
# ------------------------------------------------------
# âš™ï¸ 10. ì»¬ëŸ¼ íƒ€ì… ë¶„ë¦¬
# ------------------------------------------------------
numeric_features = data.select_dtypes(include=['int64', 'float64']).drop(columns=['Attrition']).columns.tolist()
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
```

```python
# ------------------------------------------------------
# ğŸ§ª 11. ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸ êµ¬ì„± ë° ì ìš©
# ------------------------------------------------------
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
    ]
)

X_transformed = preprocessor.fit_transform(data.drop(columns=['Attrition']))
y = data['Attrition']
```

```python
# ------------------------------------------------------
# ğŸ§¾ 12. ì „ì²˜ë¦¬ ê²°ê³¼ë¥¼ DataFrameìœ¼ë¡œ ë³µì›
# ------------------------------------------------------
num_cols = numeric_features
cat_cols = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
final_cols = np.concatenate([num_cols, cat_cols])

processed_df = pd.DataFrame(X_transformed.toarray() if hasattr(X_transformed, 'toarray') else X_transformed,
                            columns=final_cols)
processed_df['Attrition'] = y.values
```

```python
# ------------------------------------------------------
# ğŸ’¾ 13. ì €ì¥
# ------------------------------------------------------
processed_df.to_csv('data/processed_datasets.csv', index=False)
```
